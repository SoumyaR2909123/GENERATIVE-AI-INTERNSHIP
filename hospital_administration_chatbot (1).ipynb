```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

chat_df["intent"] = label_encoder.fit_transform(chat_df["intent"])

chat_df.head()
```

```python
print(label_encoder.classes_)
```

```python
chat_df["intent"].unique()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()

tokenizer.fit_on_texts(chat_df["text"])

sequences = tokenizer.texts_to_sequences(chat_df["text"])

print(sequences[:5])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

X = pad_sequences(sequences, padding='post')

print(X[:5])
print(X.shape)
```

```python
y = chat_df["intent"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2, # Changed from 0.2 to 0.4 to ensure enough samples for stratification
    random_state=42,
    stratify=y
)

print(X_train.shape)
print(X_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

vocab_size = len(tokenizer.word_index) + 1

model = Sequential([
    Embedding(input_dim=vocab_size,
              output_dim=16,
              input_length=X.shape[1]),

    LSTM(32),

    Dense(32, activation='relu'),

    Dense(len(label_encoder.classes_), activation='softmax')
])

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=8,
    validation_data=(X_test, y_test)
)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)

print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python
print("Final Training Accuracy:",
      history.history['accuracy'][-1])

print("Final Validation Accuracy:",
      history.history['val_accuracy'][-1])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

def predict_intent(text):
    seq = tokenizer.texts_to_sequences([text])

    padded = pad_sequences(
        seq,
        maxlen=X.shape[1],
        padding='post'
    )

    prediction = model.predict(padded)

    intent_index = np.argmax(prediction)

    intent = label_encoder.inverse_transform([intent_index])

    return intent[0]
```

```python
def check_appointment(appointment_id):

    result = appointments[
        appointments["appointment_id"] == appointment_id
    ]

    if len(result) == 0:
        print("Appointment not found!")

    else:
        print(result)
```

```python
def cancel_appointment(appointment_id):

    index = appointments[
        appointments["appointment_id"] == appointment_id
    ].index

    if len(index) == 0:
        print("Appointment not found!")

    else:
        appointments.loc[index, "status"] = "Cancelled"
        print("Appointment cancelled successfully!")
```

```python
def book_appointment(patient_id,
                     doctor_id,
                     appointment_date,
                     appointment_time,
                     reason):

    global appointments # Moved to the beginning of the function
    patient_id = patient_id.upper()
    doctor_id = doctor_id.upper()
    new_id = "A" + str(len(appointments) + 1).zfill(3)

    new_row = {
        "appointment_id": new_id,
        "patient_id": patient_id,
        "doctor_id": doctor_id,
        "appointment_date": appointment_date,
        "appointment_time": appointment_time,
        "reason_for_visit": reason,
        "status": "Scheduled"
    }

    appointments = pd.concat(
        [appointments, pd.DataFrame([new_row])],
        ignore_index=True
    )

    print("Appointment booked successfully!")
    print("Appointment ID:", new_id)
```

```python
%%writefile app.py

import streamlit as st
import pandas as pd
import pickle
import numpy as np

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load Model
model = load_model("hospital_chatbot_model.h5")

with open("tokenizer.pkl", "rb") as f:
    tokenizer = pickle.load(f)

with open("label_encoder.pkl", "rb") as f:
    label_encoder = pickle.load(f)

# Load Data
appointments = pd.read_csv("hospital_dataset/appointments.csv")
doctors = pd.read_csv("hospital_dataset/doctors.csv")
patients = pd.read_csv("hospital_dataset/patients.csv")
billing = pd.read_csv("hospital_dataset/billing.csv")

# Keywords
faq_keywords = [
    "parking",
    "pharmacy",
    "cafeteria",
    "wheelchair",
    "reception",
    "visitor",
    "facilities"
]

greetings = [
    "hi",
    "hello",
    "hey",
    "good morning",
    "good afternoon",
    "good evening"
]

thanks_words = [
    "thanks",
    "thank you",
    "thankyou",
    "thank u",
    "ok thanks"
]

insurance_keywords = [
    "insurance",
    "insured",
    "provider",
    "coverage",
    "claim"
]

hospital_timing_keywords = [
    "timing",
    "timings",
    "opening hours",
    "hospital open",
    "hospital timing",
    "when open",
    "working hours"
]
# Predict Intent
def predict_intent(text):

    sequence = tokenizer.texts_to_sequences([text])

    padded = pad_sequences(
        sequence,
        maxlen=20,
        padding="post"
    )

    prediction = model.predict(
        padded,
        verbose=0
    )

    predicted_class = np.argmax(prediction)

    intent = label_encoder.inverse_transform(
        [predicted_class]
    )[0]

    return intent

# Streamlit UI
st.set_page_config(
    page_title="Hospital Administration Chatbot",
    page_icon="🏥"
)

st.title("🏥 Hospital Administration Chatbot")

if "messages" not in st.session_state:
    st.session_state.messages = []

for message in st.session_state.messages:

    with st.chat_message(message["role"]):
        st.write(message["content"])

query = st.chat_input(
    "Ask a hospital-related question"
)

if query:

    query_lower = query.lower()

    st.session_state.messages.append(
        {
            "role": "user",
            "content": query
        }
    )

    if query_lower in greetings:

        response = (
            "Hello! Welcome to Hospital Administration Chatbot."
        )

    elif query_lower in thanks_words:

        response = (
            "You're welcome! Is there anything else I can help you with?"
        )

    elif any(word in query_lower for word in faq_keywords):

        response = (
            "Hospital Facilities:\n\n"
            "• Pharmacy Available\n"
            "• Parking Available\n"
            "• Cafeteria Available\n"
            "• Wheelchair Access Available"
        )

    elif any(word in query_lower for word in insurance_keywords):

        response = "Insurance Providers:\n\n"

        for provider in patients["insurance_provider"].unique():
            response += f"• {provider}\n"
    elif any(word in query_lower for word in hospital_timing_keywords):
        response = (
            "Hospital Timings:\n\n"
            "Monday - Saturday : 8:00 AM - 8:00 PM\n"
            "Sunday : 9:00 AM - 1:00 PM"
          )
    else:

        intent = predict_intent(query)

        if intent == "department_information":

            response = "Available Departments:\n\n"

            for dept in doctors["specialization"].unique():
                response += f"• {dept}\n"

        elif intent == "doctor_availability":

            response = "Available Doctors:\n\n"

            for _, row in doctors.iterrows():

                response += (
                    f"{row['doctor_id']} - "
                    f"{row['first_name']} "
                    f"{row['last_name']} "
                    f"({row['specialization']})\n"
                )

        elif intent == "hospital_timings":

            response = (
                "Hospital Timings:\n\n"
                "Monday - Saturday : 8:00 AM - 8:00 PM\n"
                "Sunday : 9:00 AM - 1:00 PM"
            )

        elif intent == "emergency_contact":

            response = (
                "Emergency Contact : +91-9876543210\n"
                "Ambulance : +91-9876543211"
            )

        elif intent == "patient_registration":

            response = (
                "Required Documents:\n\n"
                "• Aadhaar Card\n"
                "• Mobile Number\n"
                "• Insurance Details (if any)"
            )

        elif intent == "faq":

            response = (
                "Hospital Facilities:\n\n"
                "• Pharmacy Available\n"
                "• Parking Available\n"
                "• Cafeteria Available\n"
                "• Wheelchair Access Available"
            )

        elif intent == "appointment_booking":

            response = (
                "Fill the appointment details below."
            )
            st.session_state["booking_form"] = True
        elif intent == "appointment_cancellation":

            response = (
                "Enter appointment ID below."
            )
            st.session_state["cancel_form"] = True
        elif intent == "billing_inquiries":

            response = (
                "Billing Inquiry Selected.\n\n"
                "Bill lookup will be added next."
            )

        else:

            response = f"Predicted Intent: {intent}"

    st.session_state.messages.append(
        {
            "role": "assistant",
            "content": response
        }
    )

    st.rerun()
    # -----------------------------
# Booking Form
# -----------------------------

if st.session_state.get("booking_form", False):

    st.subheader("Book Appointment")
    st.info(
        "New patient? Please register first to obtain a Patient ID."
    )
    patient_name = st.text_input(
        "Patient Name"
    )

    phone_number = st.text_input(
        "Phone Number"
    )

    address = st.text_area(
        "Address"
    )

    patient_id = st.text_input("Patient ID(Leave blank if new patient)")

    doctor_option = st.selectbox(
         "Select Doctor",
         doctors["doctor_id"]
         + " - "
         + doctors["first_name"]
         + " "
         + doctors["last_name"]
         + " ("
         + doctors["specialization"]
         + ")"
    )

    appointment_date = st.date_input(
        "Appointment Date"
    )

    appointment_time = st.text_input(
        "Appointment Time"
    )

    reason = st.text_input(
        "Reason For Visit"
    )

    if st.button("Confirm Booking"):
        appointment_id = "A" + str(len(appointments) + 1).zfill(3)
        st.success(
            f"Appointment booked successfully!\n\nAppointment ID: {appointment_id}"
        )

# -----------------------------
# Cancel Form
# -----------------------------

if st.session_state.get("cancel_form", False):

    st.subheader("Cancel Appointment")

    appointment_id = st.text_input(
        "Appointment ID"
    )

    if st.button("Cancel Appointment"):

        st.success(
            f"{appointment_id} cancelled successfully!"
        )
```
